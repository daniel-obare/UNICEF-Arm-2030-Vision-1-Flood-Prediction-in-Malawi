```python
# split the data into training(2014-2015) and test set(2019)
import string
pp2015 = []
pp2019 = []
#
#
for year in flood.columns:
    if '2014' in str(year):
        pp2015.append(year)
    elif '2015' in str(year):
        pp2015.append(year)
    elif '2019' in str(year):
        pp2019.append(year)
#
#
```

```python
#extract the different columns using the lists appended
train = flood[flood.columns.difference(pp2019)]   
train.head()
```

```python
# rename the target_2015 variable as target
target = train.target_2015
```

```python
pp2019.extend(['LC_Type1_mode', 'Square_ID', 'X', 'Y','elevation'])# add xtra columns to the pp2019 list
test = flood[pp2019]
```

```python
test.head()
```

```python
# loop through all coumn names and rename them using a dictionary
# train data
#
column_2015 = {}
for column, digit in zip(pp2015, range(1, len(pp2015) + 1)):
    if 'precip' in column:
        column_2015[column] = 'week_' + str(digit) + '_precip'
#
#
#test data
column_2019 = {}
for column, digit in zip(pp2019, range(1, len(pp2019) + 1)):
    if 'precip' in column:
        column_2019[column] = 'week_' + str(digit) + '_precip'
#
#
#rename the columns using the dictionaries
train.rename(columns=column_2015, inplace=True)
test.rename(columns=column_2019, inplace=True)
```

```python
train.columns
```

```python
# align test set the same as train set
train, test = train.align(test, join = 'inner', axis = 1)
test.head()
```

## Model training

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from catboost import CatBoostRegressor
from sklearn.model_selection import cross_val_score
```

```python
test.head()
```

```python
label = train.drop( 'Square_ID', axis = 1)
rf = RandomForestRegressor()
rf1 = rf.fit(label, target)
#
# predict using the test dataset
x_test = test.drop('Square_ID', axis = 1)
rfpred=rf1.predict(x_test)
rfdf = pd.DataFrame({'Square_ID': test.Square_ID, 'target_2019': rfpred})
rfdf.head()
```

```python
# root mean square error
#
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(target, rfpred)
print('RandomForest-RMSE: ', np.sqrt(mse))
```

```python
# catboost regressor algorithm
#
cat = CatBoostRegressor(logging_level='Silent')
cat1=cat.fit(label, target)
#
# predict using the test dataset
x_test = test.drop('Square_ID', axis = 1)
catpred=cat1.predict(x_test)
catdf = pd.DataFrame({'Square_ID': test.Square_ID, 'target_2019': catpred})
catdf.head()
```

```python
# root mean square error
#
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(target, catpred)
print('CatBoost-RMSE: ', np.sqrt(mse))
```

```python
# SVR regressor algorithm
#
svr = SVR(kernel='rbf')
svr1=svr.fit(label, target)
#
# predict using the test dataset
x_test = test.drop('Square_ID', axis = 1)
svrpred=svr1.predict(x_test)
svrdf = pd.DataFrame({'Square_ID': test.Square_ID, 'target_2019': svrpred})
svrdf.head()
```

```python
# root mean square error
#
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(target, svrpred)
print('SVR-RMSE: ', np.sqrt(mse))
```

```python
# rmse for randomforest is so far the lowest
rfdf.to_csv('randomforest.csv', index = False)
```
